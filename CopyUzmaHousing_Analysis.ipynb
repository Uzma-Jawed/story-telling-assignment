Part 3: Compute Statistics

For numerical columns like:

Price | Area | Bedrooms

Compute the following using pandas:

mean() | median() | mode() | var() | std() | skew() | kurt()| min(), max()

```python
numerical_cols = ['price', 'area', 'bedrooms']
```

```python
print("Mean:\n", house_data[numerical_cols].mean())
```

```python
print("\nMedian:\n", house_data[numerical_cols].median())
```

```python
print("\nMode:\n", house_data[numerical_cols].mode().iloc[0])
```

```python
print("\nVariance:\n", house_data[numerical_cols].var())
```

```python
print("\nStandard Deviation:\n", house_data[numerical_cols].std())
```

```python
print("\nSkewness:\n", house_data[numerical_cols].skew())
```

```python
print("\nKurtosis:\n", house_data[numerical_cols].kurt())
```

```python
print("\nMinimum:\n", house_data[numerical_cols].min())
```

```python
print("\nMaximum:\n", house_data[numerical_cols].max())
```

```python
# Covariance matrix
covariance_matrix = house_data.cov(numeric_only=True)
print(covariance_matrix)
```

```python
# Correlation matrix
correlation_matrix = house_data.corr(numeric_only=True)
print(correlation_matrix)
```

```python
# üìå Visualizing Correlation
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# üìå GroupBy and Aggregation

# Average house value for each age group
house_data.groupby("price")['area'].mean().head()
```

```python
house_data.groupby("price")['area'].mean()
```

```python
# Group by 'bedrooms' and calculate mean of all numeric columns
house_data.groupby("bedrooms").mean(numeric_only=True).head()
```

```python
house_data.groupby("bedrooms")["price"].mean()
```

```python
# Mean and Median 
house_data.groupby('price')['area'].agg(['mean', 'median']).head()
```

```python
# Aggregating multiple functions on multiple columns
house_data.aggregate({
    'price': ['mean', 'sum'],
    'area': ['mean', 'max']
})
```

```python
# GroupBy with 3 Columns and Mean
grouped_mean = house_data.groupby(['price', 'area', 'bedrooms'])['stories'].mean()
grouped_mean.head()
```

```python
grouped_stats = house_data.groupby(['price', 'area', 'bedrooms'])['stories'].agg(['mean', 'max', 'min', 'count'])
grouped_stats.head()
```

```python
grouped = house_data.groupby(['bedrooms', 'airconditioning', 'furnishingstatus'])['price'].agg(['mean', 'max', 'min', 'count'])
grouped.head()
```
